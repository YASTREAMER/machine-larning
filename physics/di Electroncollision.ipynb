```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow import keras
```

```python
scalar=MinMaxScaler()
```

```python
df=pd.read_csv('D:/kaggle_data/dielectron.csv')
```

```python
df.drop(['Run','Event',],axis='columns',inplace=True)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
nan_index = df[df['M'].isnull()].index
all_columns = df.columns
#df.dropna(inplace=True)
imp_mean = IterativeImputer(random_state=0)
df = pd.DataFrame(imp_mean.fit_transform(df), columns=all_columns)
df.loc[nan_index, 'M']
```

```python
df.M.describe()
```

```python
df
```

```python
df.dtypes
```

```python
col_con=['px1','px2','py1','py2','pz1','pz2','E1','E2','pt1','pt2','eta1','eta2','phi1','phi2']
df[col_con]=scalar.fit_transform(df[col_con])
```

```python
scaled_df = scalar.fit_transform(df.to_numpy())
df = pd.DataFrame(scaled_df, columns=df.columns)
```

```python
df
```

```python
x=df.drop('M',axis='columns')
```

```python
y=df['M']
```

 x.shape

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train.columns)
```

```python
model=keras.Sequential([
    keras.layers.Dense(256,input_shape=(16,), activation="relu"),
    keras.layers.Dense(256, activation="relu"),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.01),
    loss='MeanSquaredLogarithmicError',
    metrics=['Accuracy']
)
model.fit(x_train,y_train,epochs=10)
```

```python
model = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(
            128, activation="relu", input_shape=(16,)
        ),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(10, activation="relu"),
        tf.keras.layers.Dense(1),
    ]
)
model.summary()

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='mse',
    metrics=['mse'],
)
```

```python
es = tf.keras.callbacks.EarlyStopping(patience=20)
model.fit(x_train, y_train, batch_size=32, epochs=100, callbacks=[es])
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import mean_squared_error as mse
mse(y_test, y_pred)
```
